```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
world = pd.read_csv("clean_data")
```

```python
world
```

```python
world.info()
```

```python
world.describe()
```

```python
pop_sum_by_country = pd.DataFrame(world.groupby("country")["population"].agg("sum")).reset_index().sort_values(by = "population", ascending = False)
```

```python
#The code below is also usable to create a top 10 country with the biggest population
#grouped = pd.pivot_table(world, values = "population", index = "country", aggfunc = np.sum).reset_index().sort_values(by = "population", ascending = False).head(10)
```

```python
pop_sum_by_country
```

```python
fig = plt.figure(figsize = (12,8))
ax = sns.barplot(x = pop_sum_by_country["population"][:10], y = pop_sum_by_country["country"][:10])

ax.set_xticklabels([f'{x:.0f}' + 'B' for x in ax.get_xticks()/1000000000])
plt.grid(True, axis='x')
plt.title('Top 10 Countries with the Highest Total of Population from 1960 - 2020 (in Billion)')
plt.show()
```

```python
fig = plt.figure(figsize = (12,8))
ax = sns.barplot(x = pop_sum_by_country["population"][-10:], y = pop_sum_by_country["country"][-10:])

ax.set_xticklabels([f'{x:.0f}' + 'K' for x in ax.get_xticks()/1000])
plt.grid(True, axis='x')
plt.title('Top 10 Countries with the Lowest Total of Population from 1960 - 2020 (in Million)')
plt.show()
```

```python
pop_sum_by_year = pd.DataFrame(world.groupby("year")["population"].agg("sum")).reset_index().sort_values(by = "population")
```

```python
pop_sum_by_year
```

```python
fig = plt.figure(figsize = (12,8))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(pop_sum_by_year["year"], pop_sum_by_year["population"])
ax.set_xlabel("Year")
ax.set_ylabel("Population")
ax.set_yticklabels([f'{x:.0f}' + 'M' for x in ax.get_yticks()/1000000])
ax.set_title("Total Global Population From 1960 - 2020 (in Million)")
plt.xticks(np.arange(min(pop_sum_by_year["year"]), max(pop_sum_by_year["year"])+1, 5))
plt.show()
```

Below is one example of analysis per country basis:

```python
indonesia = world[world["country"] == "Indonesia"].reset_index().drop(["country", "index"], axis = 1)
```

```python
indonesia
```

```python
fig = plt.figure(figsize = (12,8))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(indonesia["year"], indonesia["population"])
ax.set_xlabel("Year")
ax.set_ylabel("Population")
ax.set_yticklabels([f'{x:.0f}' + 'M' for x in ax.get_yticks()/1000000])
ax.set_title("Total Population of Indonesia From 1960 - 2020 (in Billion)")
plt.xticks(np.arange(min(indonesia["year"]), max(indonesia["year"])+1, 5))
plt.show()
```

Below are codes for growth rate analysis per year basis:

```python
growth_rates = []
for i in range(1, len(pop_sum_by_year["population"])):
    growth_rate = (pop_sum_by_year["population"][i] - pop_sum_by_year["population"][i-1]) / pop_sum_by_year["population"][i-1] * 100
    growth_rates.append(growth_rate)
```

```python
fig = plt.figure(figsize = (24,8))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(pop_sum_by_year["year"][1:], growth_rates, marker='o')
ax.set_xlabel("Year")
ax.set_ylabel("Population Growth Rate (%)")
ax.set_title("Total Global Population of Indonesia From 1960 - 2020 (in Billion)")
plt.xticks(np.arange(min(pop_sum_by_year["year"]), max(pop_sum_by_year["year"])+1, 2))
plt.show()
```
